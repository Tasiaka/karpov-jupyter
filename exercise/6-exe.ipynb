> Группировка (5/6)
Что выведет следующий запрос при его применении на таблице user?

```sql
SELECT country, exp_group, COUNT(id)
FROM "user"
GROUP BY country, exp_group;
```

Верно! Группировка идет по паре (country, exp_group). Соответственно, для каждой пары (country, exp_group) - именно в таком порядке - будет выдано количество пользователей. То есть, сначала будут отобраны уникальные страны, затем внутри каждой страны отберутся ее группы, и внутри каждой такой группы будет подсчитано количество пользователей.

> Join (2/2)

````python
import pandas as pd


df = pd.read_sql("""SELECT city, COUNT(us.id) FROM "user" AS us 
                FULL JOIN feed_action ON feed_action.user_id=us.id 
                FULL JOIN post ON post.id=feed_action.post_id 
                WHERE us.age = 36 and post.topic = 'covid' and action = 'view' and
                time >= '2021-11-30 00:00:00.000' and '2021-12-02 00:00:00.000' >= time 
                GROUP BY city ORDER BY COUNT(us.id) """,
    con = "postgresql://robot-startml-ro:pheiph0hahj1Vaif@"
        "postgres.lab.karpov.courses:6432/startml" )

df
````
