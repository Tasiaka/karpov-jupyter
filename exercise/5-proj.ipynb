```python
a = playstore.groupby(["category","type"]).agg({"price": "mean",  "rating": "mean","reviews": "mean"}).round(2)
a = a.rename(columns={
    "price": "mean_price",
    "rating": "mean_rating",
    "reviews": "mean_reviews"
})
a = a.fillna(0)
a["mean_rating"] = a["mean_rating"].round(1)
# a[a["mean_rating"]==0]
a.to_csv("5-project-fin.csv", sep=",")
a
```

–ö–æ–≥–¥–∞ —Ç—ã –∏—Å–ø–æ–ª—å–∑—É–µ—à—å .agg() –∏–ª–∏ pivot_table(), –º–æ–∂–Ω–æ:

–ª–∏–±–æ –ø–µ—Ä–µ–∏–º–µ–Ω–æ–≤–∞—Ç—å –ø–æ—Ç–æ–º —Å –ø–æ–º–æ—â—å—é .rename(columns={...}),

–ª–∏–±–æ —Å—Ä–∞–∑—É –∑–∞–¥–∞—Ç—å –∏–º–µ–Ω–∞ –ø—Ä—è–º–æ –≤–Ω—É—Ç—Ä–∏ .agg().

```python


# üß© –í–∞—Ä–∏–∞–Ω—Ç 1 ‚Äî –ø–µ—Ä–µ–∏–º–µ–Ω–æ–≤–∞—Ç—å –ø–æ—Å–ª–µ –∞–≥—Ä–µ–≥–∞—Ü–∏–∏
df_grouped = df.groupby(["category", "type"]).agg({
    "price": "mean",
    "rating": "mean",
    "reviews": "mean"
})

# –ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤—ã–≤–∞–µ–º –∫–æ–ª–æ–Ω–∫–∏
df_grouped = df_grouped.rename(columns={
    "price": "mean_price",
    "rating": "mean_rating",
    "reviews": "mean_reviews"
})


# ‚úÖ –≠—Ç–æ —á–∏—Ç–∞–µ–º–æ –∏ –ø–æ–Ω—è—Ç–Ω–æ, –æ—Å–æ–±–µ–Ω–Ω–æ –∫–æ–≥–¥–∞ –∫–æ–ª–æ–Ω–æ–∫ –º–Ω–æ–≥–æ.

# üß© –í–∞—Ä–∏–∞–Ω—Ç 2 ‚Äî –∑–∞–¥–∞—Ç—å –∏–º–µ–Ω–∞ —Å—Ä–∞–∑—É –≤ .agg()
df_grouped = df.groupby(["category", "type"]).agg(
    mean_price=("price", "mean"),
    mean_rating=("rating", "mean"),
    mean_reviews=("reviews", "mean")
)

```
